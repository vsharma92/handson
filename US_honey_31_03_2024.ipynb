**Which was the year when production of Honey was the highest?**

```python
df7 = df.groupby('year').sum().reset_index()
df7
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x = df7["year"],y=df7["production"])
plt.xlabel("Year(1995 - 2021)")
plt.ylabel("Total Production")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x = df7["year"],y=df7["production"])
plt.xlabel("Year(1995 - 2021)")
plt.ylabel("Total Production")
plt.show()
```

From the above plots we can conclude that the top honey production was maximum in year 2000 and minimum in year 2021

**Which was the year when total mean value of production of Honey was the highest ?**

```python
df8 = df.groupby("year")["value_of_production"].mean().reset_index()
df8
```

```python
df8.sort_values("value_of_production",ascending=False)
```

```python
plt.figure(figsize=(15,10))
sns.lineplot(x = df8["year"],y=df8["value_of_production"])
plt.xlabel("Year(1995 - 2021)")
plt.ylabel("Value_of_ Production(Dollors)")
plt.show()
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x = df8["year"],y=df8["value_of_production"])
plt.xlabel("Year(1995 - 2021)")
plt.ylabel("Value_of_ Production(Dollors)")
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.pie(df8["value_of_production"],labels=df8["year"],autopct="%1.2f%%")
plt.show()
```

From the above plots we can conclude that 2014 is the year with the highest value of production and 1999 is the year with the minumum value of production
